```python
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = pd.read_csv('output.csv')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='date',y='price', hue='item', data=data)

# Set labels and title
plt.xlabel("Date")
plt.ylabel("Price (shs)")
plt.title("Scatter Plot of Prices Over Time")

# Save the plot as an image (e.g., JPG)
plt.savefig('./static/scatter_plot.png')

```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="date",y="price", hue="item", style="item", markers=True, dashes=False)

# Set labels and title
plt.xlabel("Date")
plt.ylabel("Price (shs)")
plt.title("Line Plot of Prices Over Time")

# Save the plot as an image (e.g., JPG)
plt.savefig('./static/line_plot.png')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="date", y="price", hue="item", palette="viridis")

# Set labels and title
plt.xlabel("Date")
plt.ylabel("Price (shs)")
plt.title("Bar Chart of Prices Over Time")

# Show legend
plt.legend(title="Item", loc="upper right")
# Save the plot as an image (e.g., JPG)
plt.savefig('./static/bar_plot.png')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("husl", n_colors=len(data['item'].unique()))

# Iterate through each item and fill the area underneath the line
for i, item in enumerate(data['item'].unique()):
    subset = data[data['item'] == item]
    plt.fill_between(subset['date'], subset['price'], label=item, color=palette[i], alpha=0.4)

# Plot the line on top of the filled areas
sns.lineplot(data=data, x="date", y="price", hue="item", style="item", markers=True, dashes=False)

# Set labels and title
plt.xlabel("Date")
plt.ylabel("Price (shs)")
plt.title("Area Plot of Prices Over Time")

# Show legend
plt.legend(title="Item", loc="upper right")
# Save the plot as an image (e.g., JPG)
plt.savefig('./static/area_plot.png')
plt.show()
```
