```python
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

data = pd.read_csv('output.csv')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='date',y='price', hue='item', data=data)

# Set labels and title
plt.xlabel("Date")
plt.ylabel("Price (shs)")
plt.title("Scatter Plot of Prices Over Time")

# Save the plot as an image (e.g., JPG)
plt.savefig('./static/scatter_plot.png')

```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="date",y="price", hue="item", style="item", markers=True, dashes=False)

# Set labels and title
plt.xlabel("Date")
plt.ylabel("Price (shs)")
plt.title("Line Plot of Prices Over Time")

# Save the plot as an image (e.g., JPG)
plt.savefig('./static/line_plot.png')
```
